```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Read image
img = cv2.imread('zebra.jpg')

# RESIZING
resized = cv2.resize(img, (300, 300))

# ROTATION
rotated = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)


```

```python
# ---------------------------------------------
plt.figure(figsize=(14,10))

plt.subplot(331); plt.title("Original");
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)); plt.axis("off")

# Row 2 → Resized
plt.subplot(332); plt.title("Resized");
plt.imshow(cv2.cvtColor(resized, cv2.COLOR_BGR2RGB)); plt.axis("off")

# Row 3 → Rotated
plt.subplot(333); plt.title("Rotated");
plt.imshow(cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB)); plt.axis("off")

plt.tight_layout()
plt.show()
```
