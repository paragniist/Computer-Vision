```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("zebra.jpg")

if img is None:
    print("Error: Could not load image. Please check if 'sample.jpg' exists and the path is correct.")
else:
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    plt.imshow(img_rgb)
    plt.title("Original Image")
    plt.show()

    # Show pixel values
    print(img_rgb[100, 100])  # (R,G,B)
```

```python
plt.hist(img_rgb.ravel(), bins=256)
plt.title("Histogram")
```
